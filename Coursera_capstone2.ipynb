used Pandas read_html function to import the html table as a dataframe

```python
import pandas as pd
df2 = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")[0]
df2.head()
```

Used the first row as the index

```python
df2.columns = df2.iloc[0]
df2 = df2.reindex(df2.index.drop(0))
df2.reset_index(inplace=True)
df2.drop('index', axis=1, inplace=True)
```

Dropped the rows with ['Borough'] == 'Not assigned'

```python
df2.drop(df2.index[df2['Borough'] == 'Not assigned'], inplace = True)
```

Replaced Neighbourhood with Borough names where Neighbourhood was 'Not assigned'

```python
df2['Neighbourhood'].mask(df2['Neighbourhood']  == 'Not assigned', df2['Borough'], inplace=True)
```

Grouped the Neighbourhoods with same Postcode and Borough

```python
df3=df2.groupby(['Postcode','Borough'],group_keys=False).Neighbourhood.agg([( ', '.join)])
```

Index reset

```python
df3.reset_index()

```
